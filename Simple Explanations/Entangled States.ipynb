#### What are Product States? 

Product States are quantum states and can be written as a product product of the state of each qubit. This means that the state of each qubit of the system is independent of the state of the other qubits. In short, we multiply vectors. 

Here are some examples of Product States:

|00⟩, |01⟩, |10⟩ and |11⟩ are the Product States of two qubits.
|0⟩⊗|0⟩⊗|0⟩⊗|0⟩ is the Product State of three qubits
|+⟩⊗|-⟩ is a Bell state, which is a type of an entangled state. However, it is a Product State even if the qubits are not entangled.

Let's write it simpler: 
Here are some simple examples of product states:

* A product state where two qubits are both in state 0 is written as |00⟩.
* A product state where one of the two qubits is in state 0 and the other in state 1 is written |01⟩ or |10⟩.
* A product state where two qubits are both in state 1 is written as |11⟩.

**Entangling Gates**

Quantum Calculations usually start with a Product State. This means that the initial state of all qubits in the quantum computer is the product of the individual qubit states. (That is, the qubits are multiplied as a matrix product.) For example, if we have a quantum computer with two qubits, the initial state could be **|00⟩, |01⟩, |10⟩, or |11⟩.** A product state is a state in which the qubits are not entangled. This means that the state of each qubit is independent of the state of other qubits. On the other hand, entangled state is a state in which qubits are interrelated, meaning that the state of one qubit is affected by the state of other qubits. 


* Entangling Gates are quantum gates that connect two or more qubits and create quantum entanglement between them. Quantum entanglement is a phenomenon that connects two or more qubits in such a way that a measurement made on one of them has an immediate effect on the other qubits.


* Entangling gates are one of the most powerful features of quantum computers and are used in most quantum algorithms. For example, important quantum algorithms such as Shor's algorithm and Grover's algorithm cannot be implemented without using entangling gates.

* Despite the importance of entangling gates in quantum computing, they are difficult to implement. This is because quantum entanglement is very fragile and any error can lead to the loss of entanglement. 


* Here are some potential applications of entangling gates in quantum computing:
    - Designing new drugs and materials by simulating quantum chemistry
    - Discovery of new superconductors and other materials by simulating quantum materials science
    - Developing better artificial intelligence algorithms using quantum machine learning
    - Development of more secure communication protocols using quantum cryptography

**Entangling CX (C-NOT & Controlled-Not) Gates**

* There are many different types of entangling gates, but one of the most widely used is the **CNOT (Controlled-NOT) CX gate. The CNOT gate operates on two qubits and inverts the target qubit if the control qubit is in state 1.

* They are quantum gates that connect two qubits and create quantum entanglement between them. 
* Yes, for the CX gate, it is sufficient if the control qubit is 1. However, the target qubit must also be 1 for the CX gate to have an effect.


**Entangling CZ Gates**

Entangling CZ gates are quantum gates that connect two qubits and create quantum entanglement between them.* It also does nothing when the control qubit is in state |0, but applies a Z gate to the target qubit when the control qubit is in state |1.

* The Z gate applied to a single qubit does nothing to state |0 and gives a phase of -1 to state |1. So for the CZ gate, it only has any effect when both qubits are in state |1. No matter which one we decide to call 'control' and 'target', the effect is the same.* The Entangling CZ gate connects the phases of the two qubits. This means that two qubits can be in the same or different phases at the same time.

**What is the Difference between CX and CZ ?**

The CZ gate and the CX gate are both quantum gates that operate on two qubits and change the state of the target qubit according to the state of the control qubit. However, there are several important differences between them.

* The CX gate applies an X gate to the target qubit, while the CZ gate applies a Z gate to the target qubit. The X gate inverts the state of the qubit, i.e. it changes the state |0⟩ to |1⟩ and |1⟩ to |0⟩. The Z gate reverses the phase of the qubit, i.e. it preserves the phase of the |0⟩ state and reverses the phase of the |1⟩ state.

Let's talk about what this means. Phase is a combination of the amplitude and phase of a qubit's wave function. The phase of the qubit determines how likely the qubit is to be in the |0 state and how likely it is to be in the |1 state. The ⟨0⟩ state has a wave function with phase 1. The |1⟩ state has a wave function with phase -1. Gate Z preserves the phase of state |0⟩, but reverses the phase of state |1⟩, which causes the state of the qubit to change to state |0⟩.

* The CX gate is used to create Bell pair states, while the CZ gate is used to separate the Bell pair states. Bell pair states are states where two qubits are so interconnected that learning about the state of one qubit allows you to determine the state of the other qubit. The CX gate can be used to mix two qubits in a Bell pair state. The CZ gate, on the other hand, allows two qubits to be separated from a Bell pair state and access the states of individual qubits.

**In summary, the main differences between the CZ gate and the CX gate are:**

- The CX gate applies an X gate to the target qubit, while the CZ gate applies a Z gate to the target qubit.
- The CX gate has an effect when both qubits are in state |1⟩, while the CZ gate has an effect when only one qubit is in state |1⟩.
- The CX gate is used to create Bell-pair states, while the CZ gate is used to separate Bell-pair states.
