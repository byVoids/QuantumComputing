# 1. What is Quantum?

The behavior of both cards and compact discs can be easily explained using **classical physics:** 

* Physics is essentially a description of the world around us; we do experiments to see how the world behaves and from that we try to find out what rules the universe obeys.

But the electron orbiting an atom is a little different. In the early 1900s, scientists started to study things at the atomic scale. They found that really small things like atoms behave differently from the things we interact with in our everyday lives, and that in some cases the rules of classical physics were not quite right. The physics needed to be changed, so scientists came up with a more accurate set of rules, known as **"quantum physics".**

A natural question for computer scientists is: **"What if our bits obeyed the rules of quantum physics instead of classical physics?"** We call these bits **"qubits",** meaning **"quantum bits"**, and the computers that can work on them are called **"quantum computers".**

# 2. Exploring Cubits

Probabilities are useful when there are many possible outcomes and we do not have enough information to calculate which one will happen. For example, a roll of the dice or a coin toss. We assign a probability to each outcome and use them to calculate the likelihood of something happening. Probabilities work very well for things we usually see in the world around us, but with **"quantum"** things (like qubits) this approach fails and we need to update it to explain the behavior we see.

We can use probability amplitudes to describe quantum mechanics. Probability amplitudes are similar to normal probabilities:

* Amplitudes have a magnitude.
* Each possible outcome has a probability amplitude. The magnitude of the amplitude of this outcome tells us the probability of this outcome happening.

But amplitudes also have an extra property called **"phase".** For our purposes, phase is simply an angle that tells us which way our amplitudes are facing. You can think of phase as an angle, whereas conventional probabilities have only one magnitude, amplitudes have a direction. More specifically, an amplitude is a complex number. The consequence of phase is that when we add two of these amplitudes together, they can cancel each other out, just like positive and negative numbers. This behavior is called interference and explains all the behavior that is unique to quantum mechanics and that we don't see in classical mechanics. To find the probability of measuring an outcome, we square the magnitude of the amplitude of that outcome. This is a mathematical **"trick"** that makes everything add up nicely in the end.

**To put it in a simple way:**

Probabilities are the numbers we use to describe the likelihood of something happening. For example, if you roll a die, we can say that each number has a probability of 1/6. In quantum mechanics, probabilities work a little differently. When we measure something, we can observe it in many possible states. Each possible state is represented by a probability amplitude. Probability amplitudes are complex numbers, meaning they have both a magnitude and a phase. The magnitude of a probability amplitude tells us how likely we are to observe that outcome. The phase tells us how the outcome interacts with other possible outcomes. For example, consider a barrier that allows an electron to tunnel from one particle to another. The electron can be on both sides of the barrier, but not on both sides at the same time. There are two probability amplitudes that represent the probability of the electron being on both sides. When we add these two probability amplitudes, we get a new probability amplitude that represents the probability of the electron being on the other side of the barrier. The magnitude of this new probability amplitude depends on the magnitudes and phases of the first two probability amplitudes. Probability amplitudes are one of the most important concepts in quantum mechanics. They are used to explain many quantum phenomena such as interference, quantum entanglement and quantum teleportation. 



------------------

Amplitudes, unlike conventional probabilities, have a direction called phase as well as magnitude. Phase is an angle that indicates which direction the amplitude is pointing in. More specifically, the amplitude is a complex number. The real part represents the magnitude of the amplitude, while the imaginary part represents the phase of the amplitude. The magnitude of the amplitude represents the probability of the qubit being in that state. The phase of the amplitude represents the phase over the probability that the qubit is in that state. The phase of the amplitude is very important in quantum computing. For example, with the help of the phase, we can make decisions by comparing the state of two qubits. Also, with the help of phase, we can reverse the state of a qubit or mix the state of a qubit.

**Summary**

We learned a lot in this chapter, but the key points are the following:

* We can create pieces of objects that obey the rules of quantum mechanics, and we call them "qubits".
* These qubits can be described using probability amplitudes similar to classical probabilities but with "phase".
* These amplitudes can cancel each other (an effect called interference) and this is what causes the previously unexplained behavior.
* The best algorithms we have for simulating qubits use exponential resources with the number of qubits, so simulating large numbers of qubits is beyond the reach of classical computers.
* We have seen that quantum computers can do something that classical computers cannot: simulate the behavior of qubits! * The classical computer needs exponential resources, but the quantum computer can only perform operations and see what happens. The next interesting question is: "Can the quantum computer solve other problems that the classical computer cannot solve?". In this tutorial we will see that the answer is "yes"!
